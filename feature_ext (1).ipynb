```python
import cv2 as opencv

# Load the image
image = opencv.imread('C:\\Users\\MMP\\Downloads\\tom and jerry.jpg')

# Check if the image is loaded
if image is None:
    print("Error: Could not load image")
else:
    # Convert the image to grayscale
    gray_image = opencv.cvtColor(image, opencv.COLOR_BGR2GRAY)

    # Apply Canny edge detection
    edges = opencv.Canny(gray_image, threshold1=0, threshold2=70)

    # Apply binary thresholding on color image
    _, binary_image = opencv.threshold(image, 127, 255, opencv.THRESH_BINARY)
    _, inv_binary_image = opencv.threshold(image, 127, 255, opencv.THRESH_BINARY_INV)
    _, trunc_image = opencv.threshold(image, 150, 255, opencv.THRESH_TRUNC)
    _, to_zero_image = opencv.threshold(image, 150, 255, opencv.THRESH_TOZERO)
    _, to_zero_inv_image = opencv.threshold(image, 150, 255, opencv.THRESH_TOZERO_INV)

# Apply binary thresholding on gray image
    _, gray_binary_image = opencv.threshold(gray_image, 127, 255, opencv.THRESH_BINARY)
    _, gray_inv_binary_image = opencv.threshold(gray_image, 127, 255, opencv.THRESH_BINARY_INV)
    _, gray_trunc_image = opencv.threshold(gray_image, 150, 255, opencv.THRESH_TRUNC)
    _, gray_to_zero_image = opencv.threshold(gray_image, 150, 255, opencv.THRESH_TOZERO)
    _, gray_to_zero_inv_image = opencv.threshold(gray_image, 150, 255, opencv.THRESH_TOZERO_INV)


        
    # Display the original image and edges
    #opencv.imshow('Original Image', image)
    #opencv.imshow('Edge Detection', edges)

    # Display thresholding results
    ##opencv.imshow('Binary Image', binary_image)    
    ##opencv.imshow('Inverse Binary Image', inv_binary_image) 
    ##opencv.imshow('Truncated Image', trunc_image)   
   ## opencv.imshow('To Zero Image', to_zero_image)
   ## opencv.imshow('To Zero Inverse Image', to_zero_inv_image)

    # Wait until a key is pressed to close the window 
    #opencv.waitKey(0)

    # Close all OpenCV windows
    #opencv.destroyAllWindows()    
```

```python
image_rgb = opencv.cvtColor(image, opencv.COLOR_BGR2RGB)
binary_image_rgb = opencv.cvtColor(binary_image, opencv.COLOR_BGR2RGB)

plt.figure(figsize=(10,6))
plt.subplot(1,4,1)
plt.title('original image')
plt.imshow(image_rgb)
plt.axis('off')

plt.subplot(1,4,2)
plt.title('binary image')
plt.imshow(binary_image_rgb)

plt.subplot(1,4,3)
plt.title('original image')
plt.imshow(gray_image)
plt.axis('off')

plt.subplot(1,4,4)
plt.title('binary image')
plt.imshow(gray_binary_image)
```
